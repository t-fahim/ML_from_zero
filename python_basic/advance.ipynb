## List Comprehension
List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list

The Syntax : `newlist = [expression for item in iterable if condition == True]`

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = [x for x in fruits if x != "apple"]
print(newlist)
```

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = [x for x in fruits if "a" in x]
print(newlist)
```

## map() Function in Python
Definition and Usage : The map() function executes a specified function for each item in an iterable.

The item is sent to the function as a parameter.

Syntax : `map(function, iterables)`

```python
def my_function(n):
    return len(n)


x = map(my_function, ("apple", "banana", "cherry", "kiwi", "mango"))
print(x)
print(type(x))
print(list(x))
```

```python
def myfunc(a, b):
    return a**b


x = map(myfunc, (3, 2, 1), (1, 2, 3))

print(x)

# convert the map into a list, for readability:
print(list(x))
```

##  zip() Function Python
The zip() function takes iterables (can be zero or more),
aggregates them in a tuple, and returns

```python
a = ("John", "Charles", "Mike")
b = ("Jenny", "Christy", "Monica")
x = zip(a, b)
print(x)
print(type(x))
print(tuple(x))
```

```python
names = ["Mukesh", "Roni", "Chari", "fahim"]
ages = [24, 50, 18]
z = zip(names, ages)
print(z)
print(type(z))
print(tuple(z))
```

## Enumerate() in Python
The enumerate() function takes a collection (e.g. a tuple) and returns it as an enumerate object.

The `enumerate()` function adds a counter as the key of the enumerate object.

```python
l1 = ["eat", "sleep", "repeat"]
for e in enumerate(l1):
    print(e)
```

```python
# changing index and printing separately
for count, e in enumerate(l1, 100):
    print(count, e)
```

```python
# getting desired output from tuple
for count, ele in enumerate(l1):
    print(count)
    print(ele)
```

## Python Iterators
An iterator is an object that contains a countable number of values.

An iterator is an object that can be iterated upon, meaning that you can traverse through all the values.

```python
mytuple = ("apple", "banana", "cherry")
myit = iter(mytuple)

print(next(myit))
print(next(myit))
print(next(myit))
```

```python
mytuple = ["apple", "banana", "cherry"]
mytuple.reverse()
myit = iter(mytuple)

print(next(myit))
```
