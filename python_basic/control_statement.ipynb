# Control Statement

```python
a = 200
b = 33
if b > a:
    print("b is greater than a")
elif a == b:
    print("a and b are equal")
else:
    print("a is greater than b")
```

# iteration

```python
i = 1
while i < 6:
    print(i)
    i += 1
```

### break statement

```python
i = 1
while i < 6:
    print(i)
    if i == 3:
        break
    i += 1
```

### continue statement

```python
i = 0
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)
```

# `else` statement with `while`

```python
i = 1
while i < 6:
    print(i)
    i += 1
else:
    print("i is greater than 6")
```

# `for` statement

```python
fruits = ["apple", "mango", "banana"]
for fruit in fruits:
    print(fruit)
```

```python
for c in "banana":
    print(c)
```

# `range()` function

```python
for i in range(6):
    print(i)
```

```python
for x in range(3, 7):
    print(x)
```

```python
for x in range(1, 11, 2):
    print(x)
```

```python
# The else keyword in a for loop specifies a block of code to be executed when the loop is finished:
for x in range(6):
    print(x)
else:
    print("Finally finished!")
```

```python
for x in range(6):
    if x == 3:
        break
    print(x)
else:
    print("Finally finished!")  # not executed
```
