# Python Built in Function

```python
l = [10, 20, 30, -10]
print(len(l))
print(min(l))
print(max(l))
```

# The Math Module
Python has also a built-in module called math, which extends the list of mathematical functions.

```python
import math
```

```python
x = math.sqrt(64)
print(x)
```

```python
x = math.ceil(1.4)
y = math.floor(1.4)

print(x)  # returns 2
print(y)  # returns 1
```

```python
x = math.pi
y = math.inf
z = math.tau
print(x)
print(y)
print(z)
```

```python
# Calculate Euclidean distance
p = [3]
q = [1]
print(math.dist(p, q))

p = [3, 3]
q = [6, 12]

print(math.dist(p, q))
```
