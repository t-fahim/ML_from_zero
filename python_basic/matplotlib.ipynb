# Matplotlib Pyplot
Most of the Matplotlib utilities lies under the `pyplot` submodule, and are usually imported under the `plt alias`

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Plotting x and y points
The `plot()` function is used to draw points (markers) in a diagram.

By default, the `plot()` function draws a line from point to point.

```python
x = [1, 5, 8]
y = [1, 7, 10]
plt.plot(x, y)
plt.show()
```

# Plotting Without Line
To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.

```python
plt.plot(x, y, "o")
plt.grid(True)
plt.show()
```

You can plot as many points as you like, just make sure you have the same number of points in both axis.
- ms = marker size,
- c = color,
- mec = marker edge color,
- mfc = marker face color

```python
height = [1, 2, 3]
weight = [2, 12, 20]
plt.plot(
    height,
    weight,
    "o",
    markersize=20,
    color="green",
    markeredgecolor="red",
    markerfacecolor="yellow",
)
plt.title("Height vs Weight")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.grid(True)
plt.show()
```

### Line style : linestyle = 'dotted' or ls = ':' and linewidth = '20.5'
- 'solid' (default)	'-'
- 'dotted'	':'
- 'dashed'	'--'
- 'dashdot'	'-.'
- 'None'	'' or ' '

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {"family": "serif", "color": "blue", "size": 20}
font2 = {"family": "serif", "color": "darkred", "size": 15}

plt.title("Sports Watch Data", fontdict=font1, loc="center")  # left, right, center
plt.xlabel("Average Pulse", fontdict=font2)
plt.ylabel("Calorie Burnage", fontdict=font2)

plt.plot(x, y, linestyle="dotted")
plt.grid()
plt.show()
```

# Scatter Plot

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 10, 15, 15, 20])
plt.scatter(x, y, color="hotpink")

plt.grid()
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 10, 15, 15, 20])
plt.scatter(x, y, c="hotpink")

x = np.array([2, 4, 6, 8, 10])
y = np.array([5, 7, 10, 12, 18])
plt.scatter(x, y, color="blue")

plt.grid()
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 23, 30, 40, 50])

colors = np.array(
    [0, 10, 20, 30, 40]
)  # This array assigns a color value to each point.

plt.scatter(x, y, c=colors, cmap="rainbow")

plt.colorbar()
plt.grid()
plt.show()
```

```python
x = np.array([2, 4, 6, 8, 10])
y = np.array([5, 10, 15, 20, 25])
sizes = np.array([100, 200, 300, 400, 500])

plt.scatter(x, y, s=sizes, alpha=0.5)
plt.grid()
plt.show()
```

# Bar Chart

```python
gender = ["Male", "Female"]
count = [10, 6]
plt.bar(gender, count)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color="hotpink", width=0.5)

plt.show()
```

# Boxplot

```python
x = [1, 5, 10, 20, 3, 6, 9, 8, 7, 10, 25, 15, 100]
plt.boxplot(x)
plt.show()
```

# Pie Chart

```python
x = [10, 15, 25]
plt.pie(x)
plt.show()
```

```python
x = np.array([35, 25, 25, 15])
l = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(x, labels=l, startangle=0)
plt.legend()
plt.show()
```

```python
data = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1, 0, 0, 0]

plt.pie(data, labels=mylabels, explode=myexplode)
plt.legend(title="Four Fruits")
plt.show()
```
