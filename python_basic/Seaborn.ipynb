# Seaborn
Python's Statistical Data Visualization Library uses Matplotlib underneath to plot graphs.

It provides a high-level interface for drawing attractive and informative statistical graphics.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
print(sns.__version__)
````

# Seaborn.boxplot
``` seaborn.boxplot(data=None, *, x=None, y=None, hue=None, order=None, hue_order=None, orient=None, color=None, palette=None, saturation=0.75, fill=True, dodge='auto', width=0.8, gap=0, whis=1.5, linecolor='auto', linewidth=None, fliersize=None, hue_norm=None, native_scale=False, log_scale=None, formatter=None, legend='auto', ax=None, **kwargs) ```

````python
mart = pd.read_excel("../data_set/supermarket_sales.xlsx")
mart.head()
````

````python
sns.boxplot(data=mart, x="Total", width=0.5)
````

````python
sns.boxplot(data=mart, y="Total", width=0.2, color="green")
````

### box plt on one numeric variable by a categorical variable

````python
sns.boxplot(data=mart, x="Payment", y="Total")
````

### box plot on one numeric variable by two categorical variable using HUE attribute

````python
sns.boxplot(data=mart, x="Payment", y="Total", hue="Gender", width=0.5)
````

### add MEAN marker in the box plot using `showmeans` attrubute and change its style using `meanprops`

````python
sns.boxplot(
    data=mart,
    x="Payment",
    y="Total",
    showmeans=True,
    meanprops={"marker": "o", "markerfacecolor": "white", "markersize": "10"},
)
````

### change PALETTE, line Width

````python
sns.boxplot(
    data=mart,
    x="Payment",
    y="Total",
    showmeans=True,
    meanprops={"marker": "o", "markerfacecolor": "white", "markersize": "10"},
    palette="viridis",
    linewidth=1.5,
)
````

### box plot for each of the numeric variable in the dataframe

````python
plt.figure(figsize=(15, 10))
sns.boxplot(data=mart)
````

# Seaborn.violinplot
```seaborn.violinplot(data=None, *, x=None, y=None, hue=None, order=None, hue_order=None, orient=None, color=None, palette=None, saturation=0.75, fill=True, inner='box', split=False, width=0.8, dodge='auto', gap=0, linewidth=None, linecolor='auto', cut=2, gridsize=100, bw_method='scott', bw_adjust=1, density_norm='area', common_norm=False, hue_norm=None, formatter=None, log_scale=None, native_scale=False, legend='auto', scale=<deprecated>, scale_hue=<deprecated>, bw=<deprecated>, inner_kws=None, ax=None, **kwargs)```


````python
sns.violinplot(data=mart, y="Total")
````

### create a Violin plot on two categorical and one numeic vatiable and use split

````python
sns.violinplot(data=mart, x="Payment", y="Total", hue="Gender")
````

````python
sns.violinplot(data=mart, x="Payment", y="Total", hue="Gender", split=True)
````

### change the box in the horzontal line

````python
sns.violinplot(data=mart, y="Total", hue="Gender", inner="quarter")
````

### cut the extreme values

````python
sns.violinplot(data=mart, y="Total", hue="Gender", cut=0)
````

# seaborn.stripplot
```seaborn.stripplot(data=None, *, x=None, y=None, hue=None, order=None, hue_order=None, jitter=True, dodge=False, orient=None, color=None, palette=None, size=5, edgecolor=<default>, linewidth=0, hue_norm=None, log_scale=None, native_scale=False, formatter=None, legend='auto', ax=None, **kwargs)```




````python
sns.stripplot(data=mart, y="Total")
````

### expand markers in stripplot using jitter

````python
sns.stripplot(data=mart, y="Total", jitter=0.3)
````

### Draw line around the points using linewidth

````python
sns.stripplot(data=mart, y="Total", linewidth=0.5)
````

````python
sns.stripplot(data=mart, y="Total", x="Payment", palette="viridis")
````
