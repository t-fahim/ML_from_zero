# Central Tendency
- Mean
- Median
- Mode

# Mean

```python
speed = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]
mean = sum(speed) / len(speed)
print(mean)
```

```python
print(max(speed))
print(min(speed))
```

```python
# Use the NumPy mean() method to find the average speed
import numpy as np

speed = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]
x = np.mean(speed)
print(x)
```

# Median

```python
# Use the NumPy median() method to find the middle value
import numpy as np

speed = [99, 86, 87, 88, 86, 103, 87, 94, 78, 77, 85, 86]  # even list
x = np.median(speed)
print(x)
```

# Mode

```python
# Use the SciPy mode() method to find the number that appears the most
from scipy import stats

speed = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86, 87]
x = stats.mode(speed)
print(x)
print(x.mode)
print(x.count)
```

# Statistics — Mathematical statistics functions

```python
import statistics as st
```

```python
data = [1, 2, 2, 3, 3, 9, 9]
print(st.mean(data))
print(st.median(data))
print(st.mode(data))
```

- median_low() - Returns the lower of the two middle values when there is an even number of observations.
- median_high() - Returns the higher of the two middle values.
- median_grouped() - Estimates the median for grouped data (data that represents intervals or classes rather than exact values).
- multimode() - Returns all values that occur most frequently.

```python
l = [1, 2, 3, 4, 4, 5]
print(st.median_low(l))
print(st.median_high(l))
print(st.median_grouped(l))
print(st.multimode(l))
```

# Quantile and Percentile

```python
l = [1, 2, 2, 3, 4, 4, 5]
print(st.quantiles(l))
```

```python
# for percentile we have to use numpy
l = [10, 20, 30, 40, 50, 60, 70, 80]
print(np.percentile(l, 25))
print(np.percentile(l, 50))
print(np.percentile(l, 95))
```
